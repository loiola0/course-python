```
#importando o pandas
import pandas as pd
```

```
#para não limitar a visualização do número de colunas
pd.set_option('display.max_columns',None)
```

```
df = pd.read_csv("http://data.insideairbnb.com/brazil/rj/rio-de-janeiro/2020-06-19/data/listings.csv.gz")
```

```
#visualizar primeiras entradas
df.head()
```

```
#criando função lambda
lambda x: float(x.replace("$",""))
```

```
df['price']
```

```
#aplicando a função sobre a coluna
df['price'].apply(lambda x: float(x.replace(",","").replace("$","")))
```
